```python
#Function to get the individual player box score

        
def get_player_box_score_from_game(game_dict, player_id):
    # pull the two player lists safely
    home_players = game_dict.get("homeTeam", {}).get("players", [])
    away_players = game_dict.get("awayTeam", {}).get("players", [])

    # search both
    for p in home_players + away_players:
        if p.get("personId") == player_id:
            return p

    return None
        
```

```python
from datetime import timezone
from dateutil import parser

from nba_api.live.nba.endpoints import boxscore, scoreboard
from nba_api.stats.endpoints import leaguegamefinder
from nba_api.stats.static import players, teams


def get_today_games():
    board = scoreboard.ScoreBoard()
    games = board.games.get_dict()

    today_games = []

    for game in games:
        try:
            utc_time = parser.parse(game["gameTimeUTC"])
            local_time = utc_time.replace(tzinfo=timezone.utc).astimezone()
        except Exception:
            local_time = None

        today_games.append({
            "game_id": game.get("gameId"),
            "away_team": game.get("awayTeam", {}).get("teamName"),
            "home_team": game.get("homeTeam", {}).get("teamName"),
            "local_time": local_time,
        })

    return today_games



def print_today_games():
    games = get_today_games()

    if not games:
        print("No games found today.")
        return

    for g in games:
        if g["local_time"]:
            time_str = g["local_time"].strftime("%I:%M %p").lstrip("0")
        else:
            time_str = "TBD"

        return (f"{g['game_id']} | {g['away_team']} @ {g['home_team']} | {time_str}")
```

```python
#List of games each day

games = get_today_games()
games

print_today_games()
```

```python
#Gets players ID

def get_player_by_name(full_name):
    nba_players = players.get_players()
    for player in nba_players:
        if player["full_name"] == full_name:
            return player
    raise ValueError(f"Player not found: {full_name}")


def get_team_by_abbrev(abbrev):
    nba_teams = teams.get_teams()
    for team in nba_teams:
        if team["abbreviation"] == abbrev.upper():
            return team
    raise ValueError(f"Team abbreviation not found: {abbrev}")


get_player_by_name("Victor Wembanyama")
get_team_by_abbrev("SAS")
```

```python
#Gets teams last 5 games

def get_team_games(team_id, season=None):
    finder = leaguegamefinder.LeagueGameFinder(team_id_nullable=team_id)
    games = finder.get_data_frames()[0]

    if season:
        games = games[games.SEASON_ID.str[-4:] == str(season)]

    return games

target_team = get_team_by_abbrev("SAS")
df = get_team_games(target_team["id"], season="2025")
df.head()
```

```python
def get_last_matchup(team_id, opponent_abbrev, season=None):
    games = get_team_games(team_id, season)
    opponent_games = games[games.MATCHUP.str.contains(opponent_abbrev.upper(), na=False)]

    if opponent_games.empty:
        raise ValueError("No matchups found.")

    return opponent_games.sort_values("GAME_DATE", ascending=False).iloc[0]

# last = get_last_matchup(target_team["id"], "LAL", season="2025")
# last
```

```python
#Get player stats for a certain game ID 

def get_full_game_data(game_id):
    box = boxscore.BoxScore(game_id)
    full_box = box.game.get_dict()
    return full_box


def get_player_box_score(game_dict, player_id):
    home_players = game_dict.get("homeTeam", {}).get("players", [])
    away_players = game_dict.get("awayTeam", {}).get("players", [])

    for p in home_players:
        if p.get("personId") == player_id:
            return {
                "teamTricode": game_dict["homeTeam"].get("teamTricode"),
                "side": "home",
                "player": p,
            }

    for p in away_players:
        if p.get("personId") == player_id:
            return {
                "teamTricode": game_dict["awayTeam"].get("teamTricode"),
                "side": "away",
                "player": p,
            }

    return None


game_dict = get_full_game_data('0022500788')

player = get_player_by_name("Victor Wembanyama")
player_result = get_player_box_score(game_dict, player["id"])
player_result
```

```python
def summarize_player_box(player_result):
    if not player_result:
        return "Player not found in this game."

    player = player_result["player"]
    stats = player.get("statistics", {})

    def stat(key):
        return stats.get(key, "N/A")

    name = player.get("name") or (
        f"{player.get('firstName', '')} {player.get('familyName', '')}"
    ).strip()

    return "\n".join([
        f"{name} | {player_result['teamTricode']} ({player_result['side']})",
        f"MIN: {stat('minutes')}",
        f"PTS: {stat('points')}",
        f"REB: {stat('reboundsTotal')}",
        f"AST: {stat('assists')}",
        f"BLK: {stat('blocks')}",
        f"3PM: {stat('threePointersMade')}",
        f"+/-: {stat('plusMinusPoints')}",
    ])

print(summarize_player_box(player_result))
```

```python
def run_player_query(player_name, team_abbrev, opponent_abbrev, season=None):
    player = get_player_by_name(player_name)
    team = get_team_by_abbrev(team_abbrev)

    last_game = get_last_matchup(team["id"], opponent_abbrev, season)
    game_dict = get_full_game_data(last_game.GAME_ID)
    player_result = get_player_box_score(game_dict, player["id"])

    summary = summarize_player_box(player_result)
    print(summary)

    return {
        "player": player,
        "team": team,
        "last_matchup": last_game,
        "game_dict": game_dict,
        "player_result": player_result,
    }

run_player_query(
    player_name="Victor Wembanyama",
    team_abbrev="SAS",
    opponent_abbrev="LAL",
    season="2025"
)
```
